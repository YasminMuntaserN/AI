Creating a new sales column
What this cell does:
This cell creates a new column called sales.
The sales value is calculated by multiplying:
quantity_ordered (how many items were bought)
price_each (price of one item)
This helps us calculate the total revenue for each order.
The new column will be used in all future analysis.

```python
df['sales'] = df['quantity_ordered' ] * df['price_each']
df
```

Monthly sales analysis (best month for sales)
What this cell does:
This cell groups the data by month.
For each month, it calculates the total sales revenue using the sum.
The results are sorted from highest to lowest sales.
reset_index() is used to turn the grouped data back into a normal table.
This allows us to clearly see which month has the highest revenue.

```python
month_sales =df.groupby('month')['sales'].agg(sales_sum = 'sum').sort_values(by = 'sales_sum' , ascending = False).reset_index()
month_sales
```

Bar plot for monthly sales

Why we use a bar plot here:

Bar plots are used to compare values between categories.

Months are categories, and sales are numeric values.

Bar plots make it easy to see which month performed best.

What this cell does:

This cell creates a bar plot showing sales for each month.

The x-axis shows total sales.

The y-axis shows months.

Labels are added on the bars to display exact sales values.

The plot helps visually identify the best month for sales.

```python
fig , ax = plt.subplots(figsize=(9 ,6))

ax = sns.barplot(x = 'sales_sum' , y = 'month' , data = month_sales)
plt.title('Monthly Sales')

ax.bar_label(ax.containers[0])
plt.show()
```

Extracting city and state from address

What this cell does:

This cell extracts city and state information from the purchase_address column.

Each address contains city and state in text form.

Custom functions are used to:

Get the city name

Get the state abbreviation

A new column called city is created.

The new column combines city and state in a readable format.

This makes it easier to analyze sales by location.

```python
def get_city(address):
    return address.split(',')[1].strip(' ')

def get_state(address):
    return address.split(',')[2].strip(' ')[1]

df['city'] = df['purchase_address'].apply(lambda x : f"{get_city(x)} ({get_state(x)})")
df.head()
```

City-wise sales aggregation

What this cell does:

This cell groups the data by city.

It calculates the total sales revenue for each city.

Results are sorted from highest to lowest sales.

reset_index() is used to keep the data clean and readable.

This helps identify which city generates the most sales.

```python
city_sales =df.groupby('city')['sales'].agg(city_sales = 'sum').sort_values(by = 'city_sales' , ascending = False).reset_index()
city_sales
```

Bar plot for city-wise sales

Why we use a bar plot here:

Cities are categories, and sales are numeric.

Bar plots clearly show which city has higher sales.

What this cell does:

This cell visualizes total sales for each city.

The x-axis represents total sales.

The y-axis represents cities.

The plot makes it easy to compare sales performance between cities.

```python
fig, ax = plt.subplots(figsize=(9,6))

sns.barplot(
    x='city_sales',
    y='city',
    data=city_sales,
    ax=ax
)

plt.title('City wise Sales')

plt.show()
```

Product sales analysis

What this cell does:

This cell analyzes sales by product.

It groups products based on quantity_ordered.

The total quantity sold is calculated for each product.

Products are sorted from most sold to least sold.

This shows which products are most popular.

```python
product_sales = (
    df.groupby('product')['quantity_ordered']
      .agg(product_sales='sum')
      .sort_values(by='product_sales', ascending=False)
      .reset_index()
)

product_sales
```

Bar plot for product sales

Why we use a bar plot here:

Products are categories, and quantity ordered is numeric.

Bar plots allow easy comparison across many products.

What this cell does:

This cell displays a bar plot of products vs quantity ordered.

The x-axis shows quantity ordered.

The y-axis shows product names.

This helps quickly identify top-selling products.

```python
fig, ax = plt.subplots(figsize=(9, 6))

sns.barplot(
    x='product_sales',
    y='product',
    data=product_sales,
    ax=ax
)

plt.title('Product and Quantity Ordered')
plt.xlabel('Quantity Ordered')

plt.show()
```
