Distributions of numerical customer features (using a loop)

What this cell does:

This cell visualizes the distribution of all numerical columns in the dataset.

The dataset contains 15 numerical columns, so we need 15 plots.

Instead of writing 15 histograms manually, a loop is used to automate the process.

A figure is created with 5 rows and 3 columns, giving us 15 spaces.

Each histogram is placed in a specific position using axes[i, j].

Every histogram shows how values of one numerical feature are distributed.

This helps us understand customer behavior patterns and data spread.

Creating the figure and axes grid

What this cell does:

This cell creates one large figure with multiple plots.

The figure has:

5 rows

3 columns

The total number of plots is 15, matching the number of numerical columns.

The figure size is large to ensure all plots are readable.

A main title is added to describe the entire figure.

Looping through numerical columns to draw histograms

What this cell does:

Two nested loops are used:

The outer loop controls the rows.

The inner loop controls the columns.

numeric_columns[3*i + j] selects a different column for each subplot.

Each iteration draws one histogram in the correct position.

This approach avoids repetitive code and makes the notebook cleaner.

The result is a grid of histograms showing all numeric distributions.

```python
numeric_columns = ['account_length', 'location_code', 'user_id', 'credit_card_info_save',
       'push_status', 'add_to_wishlist', 'desktop_sessions', 'app_sessions',
       'desktop_transactions', 'total_product_detail_views',
       'session_duration', 'promotion_clicks', 'avg_order_value',
       'sale_product_views', 'discount_rate_per_visited_products',
       'product_detail_view_per_app_session', 'app_transactions',
       'add_to_cart_per_session', 'customer_service_calls', 'churn']
fig , axes =plt.subplots(5,3 , figsize = (24,24))

fig.suptitle("Distribution of numierc variables" , size = 48)

for i in range(5):
   for j in range(3):
	   sns.histplot(df[numeric_columns[3*i + j]] , ax =axes[i,j])
```

Preparing categorical columns for visualization

What this cell does:

This cell prepares categorical data for plotting.

Columns with values "yes" and "no" are converted to:

1 for yes

0 for no

This makes the data numeric, which is easier to analyze and visualize.

Only categorical (object-type) columns are modified.

```python
fig , ax =plt.subplots(1,2, figsize=(12,6))
plt.pie(df.churn.value_counts() , labels = ["Not Churn" , "churn"] , colors =["lightgreen" , "lightcoral"] , autopct = '%.0f%%' ,textprops = {'fontsize' : 14})

cp =sns.countplot(x =df['churn'] , ax =ax[0])

cp.set_xlabel("churn" , fontsize = 14 )
cp.set_ylabel("Number of customers" , fontsize = 14 )

plt.show()
```

Visualizing categorical variables related to churn

What this cell does:

This cell visualizes categorical features that may affect churn.

Bar plots are created for each categorical column.

Each plot shows how many customers fall into each category.

The plots help identify patterns related to customer churn.

Multiple plots are displayed side-by-side for easy comparison.

df.select_dtypes(include=['object']).columns.values

What this line means:

select_dtypes(include=['object'])

Selects only columns that contain text data.

These are usually categorical columns (like yes/no, city names, status).

.columns

Gets the names of those selected columns.

.values

Converts the column names into a list-like structure.


```python

categorical_columns = ["location_code","push_status","credit_card_info_save"]

for i in df.select_dtypes(include = ['object']). columns.values:
   df[i] = df[i].replace(["yes", "no"], [1, 0])

fig, ax = plt.subplots(1, len(categorical_columns), figsize = (15, 6))
for i, a in zip(categorical_columns, ax):
  temp = df[i].value_counts()
  a.set_title("Churn Counts Of " + str(i))
  sns.barplot(ax = a, x=temp.index, y=temp.values)

plt.show()
```
