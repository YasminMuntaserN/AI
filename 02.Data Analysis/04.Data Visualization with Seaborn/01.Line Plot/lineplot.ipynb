```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df =pd.read_csv('euro-daily-hist_1999_2022.csv')
df
```

```python
df =df.iloc[: , [0 , 1,4,-2]]
df.columns = ['Date' , 'Australin dollar' , 'Canadian dollar' , 'US dollar']
df
```

```python
df =pd.melt(df , id_vars = ['Date'] , value_vars = ['Date' , 'Australin dollar' , 'Canadian dollar' , 'US dollar'] , value_name='Euro rate' ,var_name ='Currency')
df['Date'] =pd.to_datetime(df['Date'])
df = df[df['Date'] >= '2022-12-01'].reset_index(drop=True)
df['Euro rate'] = pd.to_numeric(df['Euro rate'])
print(f'Currencies : {df["Currency"].unique()}\n')
print(df)
```

```python
usd =df[df['Currency'] == 'US dollar'].reset_index(drop=True)
sns.lineplot(x = 'Date' , y ='Euro rate' , data =usd)
```

```python
fig =plt.subplots(figsize=(20,5))
sns.lineplot(x = 'Date' , y ='Euro rate' , data =usd)
```

```python
from re import X
from numpy._core.defchararray import title
fig =plt.subplots(figsize=(20,5))
sns.lineplot(x = 'Date' , y ='Euro rate' , data =usd).set(title = 'Euro-USD rate' ,xlabel = 'Date' ,ylabel ='Euro')
```

```python
from matplotlib import lines
fig =plt.subplots(figsize=(20,5))
sns.lineplot(x = 'Date' , y ='Euro rate' , data =usd , linestyle='dotted' , color = "pink" , linewidth=5)
```

```python
fig =plt.subplots(figsize=(20,5))
sns.lineplot(x = 'Date' , y ='Euro rate' , data =usd , marker='*' , markerfacecolor='limegreen' , markersize = 20).set(title = 'Euro-USD rate' ,xlabel = 'Date' ,ylabel ='Euro')
```

```python
aud = df[df['Currency'] == 'Australin dollar'].reset_index(drop=True)
cad = df[df['Currency'] == 'Canadian dollar'].reset_index(drop=True)

fig = plt.subplots(figsize=(20,5))

sns.lineplot(x = 'Date' , y ='Euro rate' , data =usd)
sns.lineplot(x = 'Date' , y ='Euro rate' , data =aud)
sns.lineplot(x = 'Date' , y ='Euro rate' , data =cad)
```
