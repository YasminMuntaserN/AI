```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv('2016.csv')
df
```

```python
sns.scatterplot(data=df , x="Economy (GDP per Capita)", y="Happiness Score")
plt.show()
```

```python
sns.scatterplot(data=df , x="Economy (GDP per Capita)", y="Happiness Score")
plt.show()
```

```python
plt.figure(figsize=(15,8))

sns.scatterplot(data=df , x="Economy (GDP per Capita)", y="Happiness Score", hue="Region")
plt.show()
```

```python
grid = sns.FacetGrid(data=df , col="Region" , hue="Region" , col_wrap=5)
grid.map(plt.scatter , "Economy (GDP per Capita)" , "Health (Life Expectancy)")
grid.add_legend()
plt.show()
```
