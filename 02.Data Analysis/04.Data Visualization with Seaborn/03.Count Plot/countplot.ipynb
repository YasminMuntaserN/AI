```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =sns.load_dataset('tips')
df
```

```python
sns.countplot(x= 'sex' , data =df)
plt.show()
```

```python
sns.countplot(x= 'sex' , data =df , order=['Female' ,'Male'])
plt.show()
```

```python
sns.countplot(data = df , x='day')
plt.show()
```

```python
sns.countplot(data = df , x='day' , order=['Sat' ,'Sun' ,'Thur' , 'Fri'])
plt.show()
```

```python
sns.countplot(data = df , x='day' , order=df['day'].value_counts().index )
plt.show()
```

```python
sns.countplot(data = df , x='day' , order=df['day'].value_counts(ascending =True).index )
plt.show()
```

```python
sns.countplot(data = df , x='sex' ,hue = "smoker")
plt.show()
```

```python
ax = sns.countplot(data = df , x='sex' ,hue = "smoker")
for label in ax.containers:
  ax.bar_label(label)
plt.show()
```

```python
ax = sns.countplot(data = df , y='sex' ,hue = "smoker")
plt.show()
```

```python
ax = sns.countplot(data = df , x='sex' ,palette= "Set2")
plt.show()
```
