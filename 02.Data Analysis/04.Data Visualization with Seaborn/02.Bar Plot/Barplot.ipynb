```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.barplot(x='sex', y='survived', data=titanic)
plt.show()
```

```python
sns.barplot(x='survived', y='sex', data=titanic)
plt.show()
```

```python
sns.barplot(x='sex', y='survived', data=titanic ,palette='hls')
plt.show()
```

```python
sns.barplot(x='class', y='survived', hue='embark_town', data=titanic)
plt.show()
```

```python
sns.barplot(x='class', y='survived', hue='embark_town', errorbar= None, data=titanic)
plt.show()
```

```python
sns.barplot(x='class', y='survived', hue='embark_town', order=["Third" , "Second" , "First"], data=titanic)
plt.show()
```
