```python
import pandas as pd

df =pd.DataFrame()
print(df)
```

```python
data = [1,2,3,4,5]
df = pd.DataFrame(data)
print(df)
```

```python
df = pd.DataFrame(data , columns=["numbers"])
print(df)
```

```python
data = [['Ahmed' ,22],['Yasmin' ,21] ,['Yara' ,24]]
df = pd.DataFrame(data , columns=["Name" , "Age"])
print(df)
```

```python
df = pd.DataFrame(data ,index =["person-1","person-2","person-3"] , columns=["Name" , "Age"])
print(df)
```
