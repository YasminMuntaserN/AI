```python
import pandas as pd
```

```python
df1 =pd.DataFrame({'city' : ['A' , 'B','C' ] ,
                  'day1' : [22 ,5,28] ,
                  'day2' : [10 ,14,13] ,
                  'day3' : [22 ,15,26] ,
                  'day4' : [18 ,15,17] ,
                  'day5' : [12 ,14,28] ,
                  })
```

```python
df1.melt(id_vars=['city'])
```

```python
df1.melt(id_vars=['city'] , var_name = 'date', value_name = 'temperature')
```

```python
df1.melt(id_vars=['city'] , var_name = 'date', value_name = 'temperature').sort_values(by = 'city')
```

```python
df1.melt(id_vars=['city'] , var_name = 'date', value_name = 'temperature').sort_values(by = 'city').reset_index(drop = True)
```
