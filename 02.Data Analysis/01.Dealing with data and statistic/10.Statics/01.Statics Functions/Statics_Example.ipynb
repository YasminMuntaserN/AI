```python
import pandas as pd

df = pd.read_csv('Students_exams.csv')

df.info()
```

```python
df.rename(columns={'math score' : 'math_score' ,
                   'reading score' : 'reading_score',
                   'writing score ' : 'writing_score '
                   }, inplace=True
)
df.info()
```

```python
statics_df =pd.DataFrame(columns = ['statics' , 'math' , 'reading' , 'wirting'])
```

```python
x =df['math_score'].count()
y =df['math_score'].count()
z =df['math_score'].count()

row = {'statics' : 'count' , 'math' : x , 'reading' : y , 'wirting' : z}
# statics_df = statics_df.append(row , ignore_index=True)
statics_df = pd.concat([statics_df , pd.DataFrame.from_records([row])] , ignore_index=True)
```

```python
x =df['math_score'].min()
y =df['math_score'].min()
z =df['math_score'].min()

row = {'statics' : 'Min' , 'math' : x , 'reading' : y , 'wirting' : z}
# statics_df = statics_df.append(row , ignore_index=True)
statics_df = pd.concat([statics_df , pd.DataFrame.from_records([row])] , ignore_index=True)
```

```python
x =df['math_score'].max()
y =df['math_score'].max()
z =df['math_score'].max()

row = {'statics' : 'Max' , 'math' : x , 'reading' : y , 'wirting' : z}
# statics_df = statics_df.append(row , ignore_index=True)
statics_df = pd.concat([statics_df , pd.DataFrame.from_records([row])] , ignore_index=True)
```

```python
x =df['math_score'].median()
y =df['math_score'].median()
z =df['math_score'].median()

row = {'statics' : 'Median' , 'math' : x , 'reading' : y , 'wirting' : z}
# statics_df = statics_df.append(row , ignore_index=True)
statics_df = pd.concat([statics_df , pd.DataFrame.from_records([row])] , ignore_index=True)
```

```python
x =df['math_score'].std()
y =df['math_score'].std()
z =df['math_score'].std()

row = {'statics' : 'Std' , 'math' : x , 'reading' : y , 'wirting' : z}
# statics_df = statics_df.append(row , ignore_index=True)
statics_df = pd.concat([statics_df , pd.DataFrame.from_records([row])] , ignore_index=True)
```

```python
x =df['math_score'].mode()[0]
y =df['math_score'].mode()[0]
z =df['math_score'].mode()[0]

row = {'statics' : 'Mode' , 'math' : x , 'reading' : y , 'wirting' : z}
# statics_df = statics_df.append(row , ignore_index=True)
statics_df = pd.concat([statics_df , pd.DataFrame.from_records([row])] , ignore_index=True)
```
