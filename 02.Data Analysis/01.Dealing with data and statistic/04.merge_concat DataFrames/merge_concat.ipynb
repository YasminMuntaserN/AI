```python
import pandas as pd
```

```python
df1 =pd.DataFrame({'Column_a':[1, 2 ,3,4] , 'Column_b':['a' , 'b' , 'c' , 'd'],'Column_c':[True , True, False , True]})
```

```python
df2 =pd.DataFrame({'Column_a': [5,3,7,8] , 'Column_b':['e' , 'f' , 'g' , 'h'],'Column_c':[False , False, True , False]})
```

```python
df = pd.concat([df1 , df2])
df
```

```python
df = pd.concat([df1 , df2] , keys=['df1' , 'df2'])
df
```

```python
df = pd.concat([df1 , df2] , ignore_index = True)
df
```

```python
df = pd.concat([df1 , df2] , axis = 1)
df
```

```python
df_merege = pd.merge(df1 , df2 )
df_merege
```

```python
df_merege = pd.merge(df1 , df2 , on = 'Column_a')
df_merege
```

```python
df2.rename(columns={'Column_a' : 'new_column_a'} ,inplace=True )
df2
```

```python
df_merage = pd.merge(df1 , df2 , left_on = 'Column_a')
```

```python
df_merage = pd.merge(df1 , df2 , left_on = 'Column_a' ,  right_on= 'new_column_a')
df_merage
```
