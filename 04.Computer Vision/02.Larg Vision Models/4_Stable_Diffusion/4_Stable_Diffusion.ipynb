# What is Stable Diffusion?

Stable Diffusion is a diffusion-based generative model.

It works by:

- Starting from random noise

- Gradually removing noise

- Producing a clear image that matches the given text

Unlike older GAN-based models, Stable Diffusion uses a diffusion process, which is more stable and produces highly detailed results.

# How Does Stable Diffusion Work?

## Stage 1: Forward Diffusion (Adding Noise)

Imagine you have a clean image.

The model gradually adds random noise to it.

Step by step:

- The image becomes blurrier

- Details disappear

- Eventually it becomes pure random noise

Mathematically:
The model learns how images become noisy over time.

## Stage 2: Reverse Diffusion (Removing Noise)

Now comes the magic.

The model learns how to reverse the process:

- It starts from random noise.

- It gradually removes noise.

- At each step, it predicts what the image should look like.

Instead of restoring the original image,
it restores an image that matches the text prompt.

This is guided by a text encoder.

## Stage 3: Iterative Refinement

The noise removal process happens over many small steps (like 20–50 steps).

Each step:

- Slightly improves the image

- Makes it clearer

- Makes it more aligned with the text

Finally:
You get a realistic image that matches the prompt.

# Why Is It Called "Stable" Diffusion?

Because it is optimized to:

- Be memory efficient

- Run on GPUs with less VRAM

- Produce stable high-quality outputs

It uses a technique called Latent Diffusion, meaning:

It does diffusion in a compressed space instead of full-resolution pixels.

This makes it:

- Faster

- More efficient

- More scalable

```python
from diffusers import StableDiffusionPipeline
# StableDiffusionPipeline is a high-level interface.
# It combines:
# Text encoder
# Diffusion model
# Decoder
# Scheduler
import torch
# GPU support
# Tensor operations
# Memory optimization
```

```python
model_id = "dreamlike-art/dreamlike-photoreal-2.0"
# This is a pretrained model built on Stable Diffusion.
# It is designed to generate highly photorealistic images.
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")
```

```python
prompts = ["Majestic mountain landscape at sunset, vibrant colors, serene atmosphere",
          "Futuristic city skyline at night, neon lights, cyberpunk vibe",
          "Calm beach scene with gentle waves, sunrise, warm tones, minimalistic"]
# Prompts describe:
# Objects
# Atmosphere
# Style
# Lighting
# Mood
```
